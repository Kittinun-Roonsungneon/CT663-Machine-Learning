```python
#Reading the Dataset
Country_data=pd.read_csv("content/Country-data.csv")
Country_data
```

```python
#2.Exploratory Data Analysis
#2.1 Shape and Head of data
```

```python
#Shape of dataset
print("There are {} countries and {} features: ".format(Country_data.shape[0],Country_data.shape[1]))
#Reading the first 5 rows of the dataset
Country_data.head()
```

```python
#2.2. Data Conversion
#Converting exports, health and imports variables into specified percentages of GDPP
```

```python
Country_data['exports'] = Country_data['exports']*Country_data['gdpp']/100
Country_data['imports'] = Country_data['imports']*Country_data['gdpp']/100
Country_data['health'] = Country_data['health']*Country_data['gdpp']/100
print('After Conversion')
Country_data.head()
```

```python
# 2.3.3 Check for Null & Nan values
```

```python
# print Null and NaN values
df = pd.DataFrame(Country_data)

# à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸„à¹ˆà¸² Null
null_values = df.isnull().sum()
print("Null values per column:")
print(null_values)
# à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸„à¹ˆà¸² NaN
nan_values = df.isna().sum()
print("NaN values per column:")
print(nan_values)
```

```python
# 2.6. Describing Data
```

```python
# Checking outliers at 25%,50%,75%,90%,95% and 99%
# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸›à¸­à¸£à¹Œà¹€à¸‹à¹‡à¸™à¹„à¸—à¸¥à¹Œà¸•à¹ˆà¸²à¸‡à¹†
percentiles = [25, 50, 75, 90, 95, 99]

for p in percentiles:
    percentile_value = df['child_mort'].quantile(p / 100)
    print(f"{p}% percentile: {percentile_value}")

# à¸„à¸³à¸™à¸§à¸“à¸‚à¸­à¸šà¹€à¸‚à¸•à¸ªà¸³à¸«à¸£à¸±à¸š Outliers
Q1 = df['child_mort'].quantile(0.25)
Q3 = df['child_mort'].quantile(0.75)
IQR = Q3 - Q1

# à¹à¸ªà¸”à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹ƒà¸™à¹à¸•à¹ˆà¸¥à¸° Column
print("Data in each column:")
for column in df.columns:
    print(f"{column}: {df[column].tolist()}")
```

```python
# 2.7. Correlation Heatmap
```

```python
# à¸„à¸³à¸™à¸§à¸“ correlation matrix
df = pd.DataFrame(Country_data)
corr = df.corr()

# à¸ªà¸£à¹‰à¸²à¸‡ Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='child_mort', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 2.8. Univariate & Bivariate Analysis
# Univariate Analysis - BOXPLOT for each Numerical Variable
# à¸§à¸´à¹€à¸„à¸£à¸²à¸°à¸«à¹Œà¸«à¸² outlier à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ boxplots
# à¸•à¸±à¸§à¹à¸›à¸£à¸¡à¸µà¸„à¹ˆà¸² outlier à¸¡à¸²à¸à¸—à¸µà¹ˆà¸ªà¸¸à¸”
# à¸ˆà¸°à¸ˆà¸±à¸”à¸à¸²à¸£à¸à¸±à¸šà¸„à¹ˆà¸² outlier à¸­à¸¢à¹ˆà¸²à¸‡à¹„à¸£ à¹€à¸Šà¹ˆà¸™à¸ˆà¸°à¹€à¸à¹‡à¸šà¹„à¸§à¹‰ à¸«à¸£à¸·à¸­ à¸¥à¸š à¸­à¸­à¸ à¹€à¸žà¸£à¸²à¸°à¹€à¸«à¸•à¸¸à¹ƒà¸”
# Code
# Output
# à¸­à¸˜à¸´à¸šà¸²à¸¢à¸œà¸¥ à¸•à¸²à¸¡à¸„à¸³à¸–à¸²à¸¡
# à¸ªà¸£à¹‰à¸²à¸‡ Boxplot
# print(Country_data)
# à¸§à¸²à¸” Boxplot à¸‚à¸­à¸‡à¸—à¸¸à¸à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ
for col in df.columns:
    sns.boxplot(data=df, x=col)
# outlier_count_dict = {}
for col in df.columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outlier_count = df[col][(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    outlier_count_dict[col] = outlier_count

# max_outlier_col = max(outlier_count_dict, key=outlier_count_dict.get)
# sns.boxplot (x=Country_data['country'], y=Country_data['child_mort'])
# # à¹à¸ªà¸”à¸‡à¸à¸£à¸²à¸Ÿ
# plt.show()
```

```python
#Child Mortality Rate, Exports, Health, Imports, Income per Person, Inflation, Life Expectancy, Total Fertility, GDP per Capita
f, axes = plt.subplots(3, 3, figsize=(15, 15))

s=sns.boxplot(y=Country_data.child_mort,ax=axes[0, 0],color="#FC9803")
axes[0, 0].set_title('Child Mortality Rate')

s=sns.boxplot(y=Country_data.exports,ax=axes[0, 1],color="#FC9803")
axes[0, 1].set_title('Exports')

s=sns.boxplot(y=Country_data.health,ax=axes[0, 2],color="#FC9803")
axes[0, 2].set_title('Health')

s=sns.boxplot(y=Country_data.imports,ax=axes[2, 2],color="#FC9803")
axes[2, 2].set_title('Imports')

s=sns.boxplot(y=Country_data.income,ax=axes[1, 0],color="#FC9803")
axes[1, 0].set_title('Income')

s=sns.boxplot(y=Country_data.inflation,ax=axes[1, 1],color="#FC9803")
axes[1, 1].set_title('Inflation')

s=sns.boxplot(y=Country_data.life_expec,ax=axes[1, 2],color="#FC9803")
axes[1, 2].set_title('Life Expectancy')

s=sns.boxplot(y=Country_data.total_fer,ax=axes[2, 0],color="#FC9803")
axes[2, 0].set_title('Total Fertility')

s=sns.boxplot(y=Country_data.gdpp,ax=axes[2, 1],color="#FC9803")
axes[2, 1].set_title('GDP per Capita')

s.get_figure().savefig('boxplot subplots.png')
plt.show()
```
