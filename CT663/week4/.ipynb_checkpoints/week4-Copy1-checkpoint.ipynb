```python
# print Null and NaN values
df = pd.DataFrame(df)

# ตรวจสอบค่า Null
null_values = df.isnull().sum()
print("Null values per column:")
print(null_values)
# ตรวจสอบค่า NaN
nan_values = df.isna().sum()
print("NaN values per column:")
print(nan_values)
```

```python
# 2.6. Describing Data
```

```python
# สมมติว่า df เป็น DataFrame ที่มีข้อมูลที่ต้องการ
percentiles = [25, 50, 75, 90, 95, 99]
columns_to_check = ['year', 'Standing_area_rai', 'Fruitful_area_rai', 'Productivity_tons', 'Productivity_per_rai_kg']

# คำนวณค่าเปอร์เซ็นไทล์ต่างๆ
for p in percentiles:
    print(f"{p}% percentile values:")
    for column in columns_to_check:
        percentile_value = df[column].quantile(p / 100)
        print(f"  {column}: {percentile_value}")

# คำนวณขอบเขตสำหรับ Outliers สำหรับทุกคอลัมน์ที่ต้องการตรวจสอบ
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = max(Q1 - 1.5 * IQR, 0)  # ถ้าคำนวณออกมาเป็นค่าติดลบ ให้ใช้ค่า 0 แทน
    upper_bound = Q3 + 1.5 * IQR
    
    print(f"Outlier bounds for '{column}': lower bound = {lower_bound}, upper bound = {upper_bound}")
```

```python
# คำนวณ correlation matrix
df = pd.DataFrame(df)
corr = df.corr()

# สร้าง Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='child_mort', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 2.8. Univariate & Bivariate Analysis
# Univariate Analysis - BOXPLOT for each Numerical Variable
# วิเคราะห์หา outlier โดยใช้ boxplots
# ตัวแปรมีค่า outlier มากที่สุด
# จะจัดการกับค่า outlier อย่างไร เช่นจะเก็บไว้ หรือ ลบ ออก เพราะเหตุใด
# Code
# Output
# อธิบายผล ตามคำถาม
# สร้าง Boxplot
# print(Country_data)
# วาด Boxplot ของทุกคอลัมน์
for col in df.columns:
    sns.boxplot(data=df, x=col)
# outlier_count_dict = {}
for col in df.columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outlier_count = df[col][(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    outlier_count_dict[col] = outlier_count

# max_outlier_col = max(outlier_count_dict, key=outlier_count_dict.get)
# sns.boxplot (x=Country_data['country'], y=Country_data['child_mort'])
# # แสดงกราฟ
# plt.show()
```
